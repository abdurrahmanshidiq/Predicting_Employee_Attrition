```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import OneHotEncoder
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import cross_val_score

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import RandomizedSearchCV

from sklearn.metrics import confusion_matrix, plot_confusion_matrix, classification_report
from sklearn.metrics import f1_score, precision_score, recall_score, roc_curve, roc_auc_score, auc

sns.set_style('darkgrid')
%matplotlib inline
```

## Scaling / Standardized

```python
features = data.drop('Attrition',axis=1)
features
```

```python
robust = RobustScaler()
scaled = robust.fit_transform(features)
scaled
```

```python
features = pd.DataFrame(scaled, columns=features.columns)
features
```

```python
scaled_data = pd.concat([features, data['Attrition']], axis=1)
scaled_data
```

# Train Test Split

```python
X = scaled_data.drop('Attrition',axis=1)
y = scaled_data['Attrition']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.2, random_state=45)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
size = [len(y_train[y_train==0]),len(y_train[y_train==1])]
labels = ['Stayed', 'Left']

plt.title('Comparison Train Data (Imbalanced)', fontweight='bold')
plt.pie(size, labels=labels, autopct='%.2f%%', startangle=90)
plt.legend()
plt.show()
```

# SMOTE Oversamppling

```python
sm = SMOTE(random_state=42, sampling_strategy=1.0, k_neighbors=5)
Xsmot_train, ysmot_train = sm.fit_sample(X_train,y_train)
```

```python
Xsmot_train.shape
```

```python
size = [len(ysmot_train[ysmot_train==0]),len(ysmot_train[ysmot_train==1])]
labels = ['Stayed', 'Left']

plt.title('Comparison Train Data after SMOTE (Balanced)', fontweight='bold')
plt.pie(size, labels=labels, autopct='%.2f%%', startangle=90)
plt.legend()
plt.show()
```

# Machine Learning Modeling

## Logistic Regression

```python
logreg = LogisticRegression()
logreg.fit(Xsmot_train, ysmot_train)
```

```python
pred_logreg = logreg.predict(X_test)
proba_logreg = logreg.predict_proba(X_test)
```

```python
print(confusion_matrix(y_test, pred_logreg))
```

```python
print(plot_confusion_matrix(logreg, X_test, y_test, values_format='d'))
```

```python
print(classification_report(y_test, pred_logreg))
```

```python
cv_score_logreg = round(cross_val_score(logreg, X_test, y_test, cv = 10, scoring='accuracy').mean()*100,2)
cv_score_logreg
```

```python
cr_logreg = classification_report(y_test,pred_logreg,output_dict=True)
cr_logreg = pd.DataFrame(cr_logreg).T
cr_logreg
```

## Random Forest Classifier

```python
rfc = RandomForestClassifier()
rfc.fit(Xsmot_train, ysmot_train)
```

```python
pred_rfc = rfc.predict(X_test)
proba_rfc = rfc.predict_proba(X_test)
```

```python
print(confusion_matrix(y_test, pred_rfc))
```

```python
print(plot_confusion_matrix(rfc,X_test,y_test, values_format='d'))
```

```python
print(classification_report(y_test, pred_rfc))
```

```python
cv_score_rfc = round(cross_val_score(rfc, X_test, y_test, cv=10, scoring='accuracy').mean()*100,2)
cv_score_rfc
```

```python
cr_rfc = classification_report(y_test, pred_rfc, output_dict=True)
cr_rfc = pd.DataFrame(cr_rfc).T
cr_rfc
```

# ROC AUC

```python
proba1_logreg = proba_logreg[:,1]
fpr_logreg, tpr_logreg, threshold_logreg = roc_curve(y_test, proba1_logreg)
roc_auc_logreg = auc(fpr_logreg, tpr_logreg)

proba1_rfc = proba_rfc[:,1]
fpr_rfc, tpr_rfc, threshold_rfc = roc_curve(y_test, proba1_rfc)
roc_auc_rfc = auc(fpr_rfc, tpr_rfc)

proba1_gbc = proba_gbc[:,1]
fpr_gbc, tpr_gbc, threshold_gbc = roc_curve(y_test, proba1_gbc)
roc_auc_gbc = auc(fpr_gbc, tpr_gbc)
```

```python
plt.figure(figsize=(15,5))
plt.suptitle('ROC AUC')

plt.subplot(131)
plt.title('ROC AUC Logistic Regression')
plt.plot(fpr_logreg, tpr_logreg, 'green', label='AUC Logistic = {}'.format(round(roc_auc_logreg,2)))
plt.plot([0,1],[0,1], 'r--')
plt.xlim([0,1])
plt.ylim([0,1])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.fill_between(fpr_logreg,tpr_logreg, 0, facecolor='azure', alpha=0.5)
plt.legend(loc = 'lower right')

plt.subplot(132)
plt.title('ROC AUC Random Forest Classifier')
plt.plot(fpr_rfc, tpr_rfc, 'green', label='AUC RFC = {}'.format(round(roc_auc_rfc,2)))
plt.plot([0,1],[0,1], 'r--')
plt.xlim([0,1])
plt.ylim([0,1])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.fill_between(fpr_rfc,tpr_rfc, 0, facecolor='azure', alpha=0.5)
plt.legend(loc = 'lower right')

plt.subplot(133)
plt.title('ROC AUC Gradient Boosting Classifier')
plt.plot(fpr_gbc, tpr_gbc, 'green', label='AUC GBC = {}'.format(round(roc_auc_gbc,2)))
plt.plot([0,1],[0,1], 'r--')
plt.xlim([0,1])
plt.ylim([0,1])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.fill_between(fpr_gbc,tpr_gbc, 0, facecolor='azure', alpha=0.5)
plt.legend(loc = 'lower right')

plt.show()
```

## Evaluation Metrics

```python
acc_logerg = round(cr_logreg.iloc[2,0]*100,2)
auc_logreg = round(auc(fpr_logreg, tpr_logreg)*100,2)

acc_rfc = round(cr_rfc.iloc[2,0]*100,2)
auc_rfc = round(auc(fpr_rfc, tpr_rfc)*100,2)

acc_gbc = round(cr_gbc.iloc[2,0]*100,2)
auc_gbc = round(auc(fpr_gbc, tpr_gbc)*100,2)

all_acc = [acc_logerg, acc_rfc, acc_gbc]
all_auc = [auc_logreg, auc_rfc, auc_gbc]
all_cv_score = [cv_score_logreg, cv_score_rfc, cv_score_gbc]
```

```python
df_mtrc = pd.DataFrame([all_acc, all_auc, all_cv_score], 
                       columns=['logreg', 'rfc', 'gradboost'],
                       index=['accuracy', 'auc', 'cv_score'])
df_mtrc = df_mtrc.T
df_mtrc
```

```python
#Sort by Metric Score
df_mtrc.sort_values(['accuracy', 'auc', 'cv_score'], ascending=False)
```

```python
plt.figure(figsize=(17,5))
plt.title("Model's Evaluation Metrics Comparison", size=15, fontweight='bold')

plt.plot(df_mtrc.index, df_mtrc['accuracy'], label = 'Accuracy', color='red', marker='o', markersize=5)
plt.plot(df_mtrc.index, df_mtrc['auc'], label = 'ROC AUC', color='green', marker='o', markersize=5)
plt.plot(df_mtrc.index, df_mtrc['cv_score'], label = 'CV Score', color='blue', marker='o', markersize=5)
plt.legend(loc='lower right', prop={'size': 13})
plt.show()
```
