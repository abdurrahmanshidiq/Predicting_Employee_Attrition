```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
%matplotlib inline
```

```python
data = pd.read_csv('Human_Resources.csv')
data
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
summary = []

for col in data.columns:
    summary.append([col, data[col].dtype, data[col].isna().sum(), 
                   round(data[col].isna().sum()/len(data[col])*100,2), 
                    data[col].nunique(), list(data[col].sample(5).drop_duplicates().values)])
    
data_summary = pd.DataFrame(summary, columns=['feature', 'dtype', 'null', 'nulPct', 'unique', 'sample'])
data_summary
```

```python
data['Attrition'] = data['Attrition'].apply(lambda x : 1 if x == 'Yes' else 0)
```

```python
data['OverTime'] = data['OverTime'].apply(lambda x : 1 if x == 'Yes' else 0)
data['Over18'] = data['Over18'].apply(lambda x : 1 if x == 'Y' else 0)
```

```python
data
```

```python
data.hist(figsize=(20,30), bins=30)
plt.show()
```

```python
df_cor = data.corr()[['Attrition']].sort_values('Attrition')
df_cor
```

```python
plt.figure(figsize=(12,6))
plt.title('Features Correlation Ratio', fontweight='bold', size=13)

plt.barh(width = df_cor['Attrition'], y=df_cor.index)
plt.show()
```

```python
# Drop "over18, StandardHours, EmployeeCount"
data.drop(['EmployeeCount', 'Over18', 'StandardHours'],axis=1, inplace=True)
```
